# Segementation 연습하기

# 1. 모델 불러오기

```python
from ultralytics import YOLO

model = YOLO("C:\Potenup\DeepLearning-YOLO-Study\models\yolo11n-seg.pt")
```

# 2. 사진 예측 하기

```python
results = model("https://ultralytics.com/images/bus.jpg", save=True)
```

```python
result = results[0]
```

## 1) 마스크 확인해보기

```python
masks = result.masks.cpu() # numpy 는 cpu 에서만 계산 가능
print(len(masks)) # 탐지된 객체 수
```

```python
import matplotlib.pyplot as plt

for mask in masks.data:
    plt.imshow(mask)
    plt.axis("off")
    plt.show()
```

detection 과 동일한 boxes 결과도 확인이 가능하다
