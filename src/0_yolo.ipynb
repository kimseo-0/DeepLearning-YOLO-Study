## 성능 지표

모니터링 지표
- box_loss: 바운딩 박스 위치 예측 | 박스 위치를 얼마나 잘 맞췄는가 (낮을수록 좋음)
- cls_loss: 객체 클래스 분류 오차 | 무엇인지 맞췄는가 (낮을수록 좋음)
- dfl_loss: 박스 좌표 세부 분포 학습 오차 | 박스 테두리를 더 정밀하게 맞추는 값
- Instances: 현재 배치에서 학습된 객체 수 | 학습에 사용된 라벨 개수
- Instances(val): 검증 데이터의 객체 수 | validation에서 라벨링된 객체 개수
- Box(P): Precision (정밀도) "예측 중에 맞은 비율" (아직 낮음)
- R: Recall (재현율) | "실제 객체를 모두 잡아냈는가" (100%)
- mAP@50: IoU 0.5 기준 평균 정확도 | 예측 박스와 정답 박스가 50% 이상만 겹치면 맞은 것으로 인정
- mAP@50-95: IoU 0.5~0.95 종합 평균 정확도
